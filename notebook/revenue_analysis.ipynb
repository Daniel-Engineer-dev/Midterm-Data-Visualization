# Nhóm 1: Phân tích Xu hướng Doanh thu theo Thời gian

**Mục tiêu:**
- Biểu đồ đường thể hiện sự biến động doanh thu qua các năm/tháng.
- Xác định các chu kỳ kinh doanh đặc trưng.

**Quy tắc trực quan (Visual Rules - Slide 03):**
1. **Marks:** Sử dụng Lines (1D) để biểu diễn xu hướng thời gian.
2. **Visual Variables:** Sử dụng Position (x, y) trên hệ tọa độ chung (most accurate channel for quantitative data).
3. **Expressiveness:** Chỉ hiển thị thông tin doanh thu và thời gian, tránh các chi tiết gây nhiễu.
4. **Effectiveness:** Nhấn mạnh các điểm đỉnh (peak) và đáy (trough) để nhận diện chu kỳ.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates

# Set style
plt.style.use('seaborn-v0_8-muted')
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 12
```

## 1. Load Data

```python
df = pd.read_csv('../data/ecommerce_sales_cleaned.csv')
df['order_date'] = pd.to_datetime(df['order_date'])
df.head()
```

## 2. Xử lý dữ liệu thời gian
Để phân tích xu hướng, chúng ta sẽ gộp doanh thu theo Tháng.

```python
# Tạo cột Year-Month để gom nhóm
df['month_year'] = df['order_date'].dt.to_period('M')
monthly_revenue = df.groupby('month_year')['total_amount'].sum().reset_index()
monthly_revenue['month_year_dt'] = monthly_revenue['month_year'].dt.to_timestamp()
monthly_revenue.head()
```

## 3. Trực quan hóa Xu hướng Doanh thu

Chúng ta sử dụng biểu đồ đường (Line chart) với trục X là thời gian và trục Y là doanh thu (total_amount). Theo quy tắc Slide 03, **Vị trí (Position)** là kênh hiệu quả nhất để biểu diễn dữ liệu định lượng.


```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=monthly_revenue, x='month_year_dt', y='total_amount', marker='o', linewidth=2.5, color='#2c3e50')

# Thêm các vùng highlight chu kỳ (nếu thấy rõ)
# Ví dụ: Highlight Q4 (Cuối năm thường tăng mạnh)
plt.axvspan(pd.Timestamp('2023-10-01'), pd.Timestamp('2023-12-31'), color='green', alpha=0.1, label='Holiday Season 2023')
plt.axvspan(pd.Timestamp('2024-10-01'), pd.Timestamp('2024-12-31'), color='green', alpha=0.1, label='Holiday Season 2024')

plt.title('Xu hướng Doanh thu hàng tháng (2023 - 2025)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tháng/Năm', fontsize=12)
plt.ylabel('Tổng Doanh thu (USD)', fontsize=12)

# Định dạng trục X
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%m/%Y'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=2))
plt.xticks(rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
```

## 4. Phân tích chu kỳ kinh doanh

Bây giờ ta sẽ xem xét doanh thu trung bình theo tháng để xác định chu kỳ lặp lại hàng năm.

```python
seasonal_revenue = df.groupby('order_month')['total_amount'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=seasonal_revenue, x='order_month', y='total_amount', hue='order_month', palette='viridis', legend=False)

plt.title('Doanh thu trung bình theo Tháng trong năm', fontsize=14, fontweight='bold')
plt.xlabel('Tháng', fontsize=12)
plt.ylabel('Doanh thu trung bình', fontsize=12)
plt.xticks(range(0, 12), ['Th1', 'Th2', 'Th3', 'Th4', 'Th5', 'Th6', 'Th7', 'Th8', 'Th9', 'Th10', 'Th11', 'Th12'])
plt.tight_layout()
plt.show()
```
